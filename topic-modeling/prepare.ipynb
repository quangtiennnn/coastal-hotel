# Prepare — Extract en/vi Reviews

Reads `data/agoda-reviews.csv`, keeps only rows where `language` is `en` or `vi`,
and saves the result to `data/agoda-reviews-en-vi.csv`.

```python
import pandas as pd
from pathlib import Path

INPUT_PATH  = Path("../data/agoda-reviews.csv")
OUTPUT_PATH = Path("../data/agoda-reviews-en-vi.csv")

df = pd.read_csv(INPUT_PATH, encoding="utf-8-sig")
print(f"Total rows      : {len(df):,}")
print(f"Language counts :{df['language'].value_counts()}")

df_filtered = df[df["language"].isin(["en", "vi"])].reset_index(drop=True)
print(f"After filter    : {len(df_filtered):,} rows")
print(df_filtered["language"].value_counts())

df_filtered.to_csv(OUTPUT_PATH, index=False, encoding="utf-8-sig")
print(f"Saved -> {OUTPUT_PATH.resolve()}")
```
