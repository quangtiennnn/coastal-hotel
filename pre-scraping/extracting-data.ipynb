```python
result = {}

hotel_el = soup.select_one('.DUwDvf.lfPIob')
if hotel_el:
    result["hotel_name"] = hotel_el.text.strip()

# Add rating and no_reviews from '4.2(174)' format or just '4.2'
rating_el = soup.select_one('.F7nice')
if rating_el:
    rating_text = rating_el.text.strip()
    match = re.match(r'(\d+\.?\d*)\((\d+)\)', rating_text)
    if match:
        result["rating"] = float(match.group(1))
        result["no_reviews"] = int(match.group(2))
    else:
        # If no parentheses, assume just rating, set no_reviews to 0
        rating_match = re.match(r'(\d+\.?\d*)', rating_text)
        if rating_match:
            result["rating"] = float(rating_match.group(1))
            result["no_reviews"] = 0

# Add accommodation
accom_el = soup.select_one('.mgr77e')
if accom_el:
    result["accommodation"] = accom_el.text.strip()[1:]

for el in soup.select("[data-item-id][aria-label]"):
    raw_id = el.get("data-item-id", "").strip()
    aria = el.get("aria-label", "").strip()
    if not raw_id:
        continue

    if raw_id.split(":")[0] == "phone":
        key = "phone"
    else:
        key = raw_id  # or key = normalized[1].strip() if you want only suffix

    # use aria label as value (or use text if you need)
    result[key] = aria

# Add facilities from .gSamH elements
result["facilities"] = [el.text.strip() for el in soup.select('.gSamH') if el.text.strip()]

# Restructure into metadata and informations
metadata_keys = ["hotel_name", "rating", "no_reviews", "accommodation"]
metadata = {k: result[k] for k in metadata_keys if k in result}
metadata["scrape_time"] = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
informations = {k: v for k, v in result.items() if k not in metadata_keys}


final_result = {
    "metadata": metadata,
    "informations": informations
}

# # Save to JSON
# import json
# from pathlib import Path

# output_path = Path("pre-scraping/extracted_data.json")
# output_path.write_text(json.dumps(final_result, ensure_ascii=False, indent=2), encoding="utf-8")
# print("Saved to", output_path)
final_result
```
